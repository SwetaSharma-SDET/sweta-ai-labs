```python
from typing import TypedDict, Optional
from langgraph.graph import StateGraph, START, END

# class MyState extends TypedDict
class MyState(TypedDict):
    # name: Optional[str]
    # age: Optional[int]
    label: str
    order: int
    response: str

    #Set grade to age - 5 for each node



def scan_book(state: MyState) -> MyState:
    
    print("scan_book ran")
    print(f"state at the beginning of node ScanBook: {state}") # state =  {"label": "START NODE", "order": 1, "response": ""}
    
    state["label"] = "Node ScanBook"
    state["order"] = state["order"] + 1
    state["response"] = " ScanBook complete"
    
    print(f"state at the end of node ScanBook: {state}") # state =  {"label": "START NODE", "order": 1, "response": ""}
    return state # state = {"label": "Node A", "order": 2, "response": "Hello from Node A"}

def give_recommendation(state: MyState) -> MyState:
    print("RecommendRelatedBook ran")
    print(f"state at the beginning of node RecommendRelatedBook: {state}") # state =  {"label": "Node A", "order": 2, "response": "Hello from Node A"}
    
    state["label"] = "Node RecommendRelatedBook"
    state["order"] = state["order"] + 1
    state["response"] = "RecommendRelatedBook complete"
    
    state["response"] = "RecommendRelatedBook complete"
    print(f"state at the end of node RecommendRelatedBook : {state}") # state =  {"label": "Node B", "order": 3, "response": "Hello from Node B"}
    # print(f"node_b: {state["label"]} is {state["order"]} years old") 
    return state


builder = StateGraph(MyState)
builder.add_node("ScanBook", scan_book)
builder.add_node("RecommendRelatedBook", give_recommendation)

builder.add_edge(START, "ScanBook")
builder.add_edge("ScanBook", "RecommendRelatedBook")
builder.add_edge("RecommendRelatedBook", END)

graph = builder.compile()
```

```python

print(graph.get_graph().draw_ascii())

initial_state={"label": "START NODE", "order": 1, "response": ""}

result = graph.invoke(initial_state)
print(result)
```
