## Build a simple sequential graph using Langgraph

```python
from langgraph.graph import StateGraph , START, END
```

```python
def node_a_agemt(dict):
    print("ImageRecognitionAgent is running")
    return{}

def node_b_agent(dict):
    print("RecommendationAgent is running")
    return{}

builder = StateGraph(dict)

builder.add_node("ImageRecognitionAgent", node_a_agemt)
builder.add_node("RecommendationAgent", node_b_agent)

name_agent_a="ImageRecognitionAgent"
name_agent_b="RecommendationAgent"

builder.add_edge(START, name_agent_a)
builder.add_edge(name_agent_a, name_agent_b)
builder.add_edge(name_agent_b, END)

graph=builder.compile()


```
