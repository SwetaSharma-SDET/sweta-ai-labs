```python
import os
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv(override=True, dotenv_path="../.env.local")
api_key=os.getenv("OPENAI_API_KEY")
print(api_key[-5:])
client=OpenAI(api_key=api_key)

```

```python
while True:

    details = input("You: ")

    if details.lower() in ["exit", "bye"]:
        break

    else:

        response = client.chat.completions.create(
            model="gpt-4.1",
            messages=[
                {
                    "role": "system",
                    "content": """
You are a helpful real estate listing assistant .

Provide:
1. A structured JSON output
2. A short marketing blurb

Use null or empty strings when information is unavailable
Use Expert broker tone
Must to have: Ask Clarifying questions at the end when important information is missing or null for Listing the property
Do not include Clarifying questions in json response
Do not hallucinate missing data
"""
                },
                {
                    "role": "user",
                    "content": details
                }
            ]
        )

        print(response.choices[0].message.content)
```
