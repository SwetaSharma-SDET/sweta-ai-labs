```python
import ollama 

client = ollama.Client()

```

```python

user_input = input("What is your question? ")
prompt = user_input
response = client.chat(
    model="llama3",
    messages=[
        {"role": "system", "content": "You are a helpful assistant. Please answer clearly."}, # Single line with double quotes, if we have single line system prompt. If we have multi line system prompt, then we can use triple quotes.

        {"role": "user", "content": prompt}
    ]
)
print (response.message.content)
```
