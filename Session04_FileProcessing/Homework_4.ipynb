## Q1. Two Sum Variant
Given a list of integers and a target value, return the indices of two numbers whose sum equals the target.
Example
nums = [2, 7, 11, 15]
target = 9

Output:
[0, 1]


```python
from typing import List, Tuple

def find_indices(nums: List[int], target: int ) -> Tuple[int, int]:
    num_dict={}

    for i in range(len(nums)):
        diff=target-nums[i]
        if diff in num_dict:
            return num_dict[diff], i
        num_dict[nums[i]]=i
    return -1,-1

nums = [2, 9, 11, 7]
target = 9

print(find_indices(nums, target))
```

Q2. Longest Substring Without Repeating Characters
Given a string, find the length of the longest substring without repeating characters.
Example
Input:
"abcabcbb"

Output:
3
Explanation:
 "abc" is the longest substring.

```python
def find_len_longest_substring(input: str) -> int:
    unique_letter=set()
    right=0
    left=0
    max_len=0
    cur_len=0

    # put letter on right index in set if set already contains them move left index by one keep incrementing the right index

    for right in range(len(input)):
        while input[right] in unique_letter:
            unique_letter.remove(input[left])
            left+=1
        unique_letter.add(input[right])
        cur_len=right-left+1
        max_len=max(max_len,cur_len)
    return max_len


print(find_len_longest_substring("abcabcbb"))
print(find_len_longest_substring("abcdabcbb"))
print(find_len_longest_substring("a0989821aaaaa"))





```

Q3. Group Anagrams
Given a list of strings, group all anagrams together.
Example
Input:
["eat", "tea", "tan", "ate", "nat", "bat"]

Output:
[
 ["eat", "tea", "ate"],
 ["tan", "nat"],
 ["bat"]
]

```python
from typing import List
from collections import defaultdict

def list_anagrams(words : List[str]) -> List[List[str]]:
    anagrams_dict=defaultdict(list)

    for word in words:
        key = "".join(sorted(word))
        anagrams_dict[key].append(word)
    return list(anagrams_dict.values())
    

print(list_anagrams(["eat", "tea", "tan", "ate", "nat", "bat", "  tan"]))


        
                        
                      

                    
        
```

Q4. Valid Parentheses
Determine if a string containing ()[]{} is valid.
Example
Input:
"({[]})"

Output:
True


```python
def is_valid_string(string :str) -> bool:
    stack = []
    # if string is empty then return true
    if len(string) ==0:
        return True
    for i in range(len(string)):
        cur_in=string[i]
        if string[i] in "({[":
            stack.append(string[i])
        elif string[i] in ")}]":
            if not stack:
                return False
            last_in=stack.pop()
            if cur_in==")" and last_in!="(" or cur_in=="}" and last_in!="{" or cur_in=="]" and last_in!="[":
                return False
    return len(stack)==0

print(is_valid_string("({[{]})"))
```

Q5. Merge Intervals
Given a list of intervals, merge overlapping intervals.
Example
Input:
[[1,3],[2,6],[8,10],[15,18]]

Output:
[[1,6],[8,10],[15,18]]
Concepts Covered
Sorting
Arrays
Interval problems

```python
def merge_intervals(intervals: List[List[int]]) -> List[List[int]]:
    intervals.sort()
    output=[intervals[0]]

    for i in range( 1,len(intervals)):
        # if number at index 0 in interval list is smaller than number at index 1 in output list than put number at index 1 in interval list in outout index 1 
        last_end=output[-1][1]
        cur_start=intervals[i][0]
        cur_end=intervals[i][1]
        if cur_start<=last_end:
            output[-1][1]=max(last_end , cur_end)
        else:
            output.append(intervals[i])
    return output

print(merge_intervals([[1,3],[2,6],[8,10],[15,18]]))
```

Q6. Top K Frequent Elements
Given an integer array and an integer k, return the k most frequent elements.
Example
Input:
nums = [1,1,1,2,2,3]
k = 2

Output:
[1,2]


```python
import heapq
def find_kth_largest_element(nums: List[int], n: int) -> List[int]:
    heap=[]

    for num in nums:
        heapq.heappush(heap,num)
        if len(heap)>n:
            heapq.heappop(heap)
    return heap

print(find_kth_largest_element([1,1,1,2,2,3], 2))
```
