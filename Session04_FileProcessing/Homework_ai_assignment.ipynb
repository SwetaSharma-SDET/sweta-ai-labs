Event Extraction from Image (Flyer or Poster)
Pick an image flyer (conference, concert, workshop, meetup, webinar, etc.).
Use only OpenAI GPT-4.1 to extract:
Event name
Date and time
Location
Organizer
Event summary (2–3 lines)
Deliverable
Submit:
event_openai.json
The image used
Example:
{
 "event_name": "AI Innovation Summit 2026",
 "date_time": "June 15, 2026, 10:00 AM",
 "location": "San Francisco Convention Center",
 "organizer": "TechWorld AI",
 "summary": "A conference focused on the future of Generative AI, AI agents, and enterprise automation."
}


```python
import os
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv(override=True, dotenv_path="../.env.local")
api_key=os.getenv("OPENAI_API_KEY")
print(api_key[-5:])
client=OpenAI(api_key=api_key)
```

```python
import base64

image_path = "data/concert.jpg" 
```

```python
with open(image_path, "rb") as f:
    image1_base64 = base64.b64encode(f.read()).decode("utf-8")
image1_base64
```

```python
mime_type = "image/png"

response = client.chat.completions.create(
    model="gpt-4.1",
    messages=[
        {
            "role": "system",
            "content": "You are technical assistant who helps to extract and summarize information from event advertisement or flyers."
        },
        
        {
            "role": "user",
            "content": [
                {"type": "text", "text": '''
                 Extract as many fields, such as Event name ,Date and time, Location, Organizer, Event summary (2–3 lines),Deliverable "
                 number, data, due date etc, Invoice product list from this image as a JSON object:
                 '''},


                {"type": "image_url", "image_url": {"url": f"data:{mime_type};base64,{image1_base64}"}}
            ]
        }
    ],
)

print("🧾 Extracted Info from Image:\n")
print(response.choices[0].message.content)
```
