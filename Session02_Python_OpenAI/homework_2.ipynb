
Part 6: Practice Problem – Student Scores
Given:
python

scores = [78, 85, 62, 90, 56, 99, 74]

Write code to:
Calculate and print the average score.


Print all scores greater than 80.


Replace scores below 60 with the word "Fail".



```python
scores = [78, 85, 62, 90, 56, 99, 74]

#Calculate and print the average score.

print(" average score is : " , sum(scores)/len(scores))

#Print all scores greater than 80.

for score in scores:
    if score>80:
        print(score)

#Replace scores below 60 with the word "Fail".
for i in range(len(scores)):
    if scores[i]<60:
        scores[i]="fail"

print("updated scores:" , scores)



```

Part 8: Moderation
Goal:
Understand how input and output moderation works using the OpenAI Moderation API.
Setup:
Use the GPT moderation code we worked on in class.
Tasks
Create a chatbot with input moderation
Before processing the user’s message:
Send the input to the OpenAI Moderation API
Check whether the content is flagged
If the input is:
Not flagged → process the request normally using GPT
Flagged → block the request and display a warning message
Example:
“Your input violates content safety guidelines.”

Add output moderation
After the GPT model generates a response:
Send the output text through the Moderation API
Check whether the response contains toxic or unsafe content
If the output is:
Safe → display the response to the user
Flagged → replace the response with a safe fallback message
Example:
“The generated response was blocked due to unsafe content.”


```python
from prompt_toolkit import prompt

while True:
    my_input= input(" Ask Question :  ")
    print( " You: " , my_input)

    if (my_input.lower() in ["exit", "bye", "Quit"]):
        print(" Good Bye !!")
        break
    else:
        # Moderation check
        mod_result=client.moderations.create(
            model="omni-moderation-latest",
            input=my_input
        )

        print(f"Moderattion result: {mod_result}")
        print(f"is flagged?: {mod_result.results[0].flagged}")
        is_content_flagged=mod_result.results[0].flagged

        if (is_content_flagged):
            print("Your input violates content safety guidelines. Please ask a different question")
            continue
        else:
            my_response=client.chat.completions.create(
                model="gpt-4.1-nano",
                messages=[
                    {"role": "system", "content": "you are a helpful assistant, provide a short and logical answer."},
                    {"role": "user", "content": my_input}
                ]
            )
        chat_response=my_response.choices[0].message.content
        
        if ("stupid","hurt" in chat_response):
            print("The generated response was blocked due to unsafe content.")
        else:
            print(my_response.choices[0].message.content)
          
```
