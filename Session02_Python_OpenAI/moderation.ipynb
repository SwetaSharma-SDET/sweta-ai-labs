```python
import os
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv(override=True, dotenv_path="../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")

print ( my_api_key[-10:])
```

```python
client=OpenAI(api_key=my_api_key)
question=input(" what is your question")
```

```python
# Step 1: Run moderation check
mod_result = client.moderations.create(
    model="omni-moderation-latest",
    input=question
)

print(f"Moderation result: {mod_result}")
print(f"mod_result.results[0].flagged result: {mod_result.results[0].flagged}")
```

```python
from prompt_toolkit import prompt


client = OpenAI()
while (True):
    question = input("What is your question? ")
    print("You asked: " + question)


    if question.lower() == "exit":
        print("Goodbye!")
        break
    else:
        
        # Step 1: Run moderation check
        mod_result = client.moderations.create(
            model="omni-moderation-latest",
            input=question
        )

        print(f"Moderation result: {mod_result}")
        print(f"mod_result.results[0].flagged result: {mod_result.results[0].flagged}")

        # Step 2: Generate response if not flagged
        if (mod_result.results[0].flagged):
                print("The content of the question is flagged by moderation. Please ask a different question.")
                continue
        else:
            response = client.chat.completions.create(
                model="gpt-4.1-nano", #gpt-5-mini
                messages=[
                    {"role": "system", "content": '''You are an AI assistant. 
                            Please answer the question as best as you can. 
                            If you don't know the answer, say you don't know.'''},
                    {"role": "user", "content": question}
                ]
            )
            text_answer = response.choices[0].message.content
            print (text_answer)
```
