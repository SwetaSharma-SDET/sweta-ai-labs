1. List Comprehension
Create a list of squares of all even numbers from 1 to 20.
Example:
[4, 16, 36, ...]

```python
from typing import List
def number_squares() -> List[int]:
    num = 1
    squares=[]

    while num<=20:
        if num%2==0:
            squares.append(num**2)
        num+=1
    return squares

print(number_squares())
```

2. Dictionary Practice
Given a list of student names and scores:
Create a dictionary
Print students who scored above 80
Example Input:
students = [("Alice", 92), ("Bob", 76), ("Charlie", 88)]

```python
def students_scores(students: List[tuple]) -> dict[str,int]:
    student_dict={}

    for i in range(len(students)):
        student_name=students[i][0]
        student_score=students[i][1]
        # add student name as a key in student_dict
        # add student score as a key value in student_dict
        student_dict[student_name]=student_score
        if student_score >80:
            print(f"{student_name} score {student_score}")
    return student_dict

students = [("Alice", 92), ("Bob", 76), ("Charlie", 88)]
students_scores(students)
```

3. Functions
Write a Python function:
get_top_k_docs(query, embeddings, k)

The function should:
Compute cosine similarity
Return the top-k most similar documents

```python
from sentence_transformers import SentenceTransformer
import numpy as np
from typing import List

def get_top_k_docs(query: str, docs : List[str], k :int) -> List[str]:
    model = SentenceTransformer("all-MiniLM-L6-v2")
    docs_vector= model.encode(docs)
    query_vector=model.encode([query])
    similarity=model.similarity(docs_vector,query_vector)[0]
    similarity=np.asarray(similarity).squeeze
    print(similarity)
    top_k_matches_index=np.argsort(similarity)[::-1][:k]
    print(top_k_matches_index)
    top_k_matches= [docs[i] for i in top_k_matches_index]
    return top_k_matches

company_doc=[
        "Employees are compensated on a bi-weekly basis through direct deposit.",       
        "Employees must submit a leave request for approval.", 
        "Company internet must be used for work-related tasks only.",
        "Company internet is a broadband internet.",
        "Employees can take an hour break.",
        "Interact with each employee with Respect"
]

print(get_top_k_docs("internet policy", company_doc, 3))



```
