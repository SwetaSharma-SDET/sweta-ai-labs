```python
import os
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path="../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")
if not my_api_key:
    raise ValueError("API key is not found")
print(my_api_key[-5:])
```

```python
from openai import OpenAI
client = OpenAI(api_key=my_api_key)
```

```python
prompt = "Tell me something about Lunar Year"
try:
    response = client.chat.completions.create(
        model="gpt-4.1-nano",
        messages=[
            {"role": "system" , "content" : "You are a poet , Answer user's question like you a poem"},
            {"role": "user" , "content" : prompt}  
                  ]            
    )
    print (response)
    print (response.choices[0].message.content)
except Exception as e:
    print (f"Error occured : {e}")
```

```python
prompt = "Tell me something about Lunar Year"
try:
    respose = client.chat.completions.create(
        model="gpt-5-nano",
        messages=[
            {"role": "system" , "content" : "You are a poet , Answer user's question like you a poem"},
            {"role": "user" , "content" : prompt}  
                  ]            
    )
    print (respose)
    print (respose.choices[0].message.content)
except Exception as e:
    print (f"Error occured : {e}")
```

```python
while True:
    prompt = input("User: ")

    if prompt.lower() in ("exit", "quit", "bye"):
        print("Chatbot: Goodbye!")
        break

    try:
        response = client.chat.completions.create(
            model="gpt-4.1-mini",
            messages=[
                {"role": "system", "content": "You are a helpful teacher. Explain clearly."},
                {"role": "user", "content": prompt}
            ]
        )

        print("Chatbot:", response.choices[0].message.content)

    except Exception as e:
        print(f"Error occurred: {e}")
```
