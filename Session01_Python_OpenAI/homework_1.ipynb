Question 1: Reverse a String 
Write a function that reverses a given string without using built in functions [:: -1]

```python
def reverse_string(s):
    # covert string to list since string is immutable 
    my_list=list(s)

    left=0
    right=len(my_list)-1

    while left<right:
        my_list[left], my_list[right] = my_list[right], my_list[left]
        left += 1
        right -= 1
    
    # convert the list to string
    return ''.join(my_list)

print(reverse_string("Hello"))
print(reverse_string(" h e l l o"))
```

Question 2: Write a function that checks if a string is a palindrome. Ignore case and spaces.

```python
def is_palindrome(s): 
    left =0
    right=len(s)-1
    s = s.lower()

    while left<right:
        # Keep increasing the left index pointer if space is found
        while left<right and s[left]==" ":
                left +=1

        # Keep decreasing the right index pointer if space is found
        while left<right and s[right]==" ":
                right -=1
    
        # Compare the character at left and right indexes
        if s[left] != s[right]:
            return False


        left +=1
        right -=1

    return True

print(is_palindrome("121"))
print(is_palindrome("    122 2 1 "))
```

Question 3 : Find the Missing Number
Example: [1,2,4,5] -> Missing : 3

```python
def find_missing_number(nums):
        
        # max number in the list
        n = len(nums) + 1

        number = 1
        e_sum = 0

        while number <= n:
                e_sum = e_sum +number
                number +=1
        
        a_sum=0
        i=0
        l=len(nums)-1

        while i<=l:
                a_sum=a_sum + nums[i]
                i +=1
        
        missing_num = e_sum - a_sum

        return missing_num

print(find_missing_number([1,2,4,5]))

# doesn't work in below case, this solution work when there is only one missing number and no duplicates
print(find_missing_number([1,2,4,5,5]))
```

Question 4 : Count Word Frequency
 Given a sentence, return a dictionary with word counts.
 Example: "AI makes AI easier" → { "AI": 2, "makes": 1, "easier": 1 }.

```python
def word_count(sentence) -> dict:
    words=sentence.split()
    word_count={}

    for i in range (len(words)):
        word=words[i]
        if word in word_count:
         word_count[word]=word_count[word] +1
        else:
            word_count[word]=1
    return word_count

print(word_count("AI makes AI easier"))
print(word_count("AI makes AI easier via AI"))
```

Question 5: FizzBuzz
 Print numbers from 1 to 50. For multiples of 3, print "Fizz", for multiples of 5 print "Buzz", and for multiples of both, print "FizzBuzz"

```python
def print_fizzBizz():
    num=1

    while num<=50:
        if num%3==0:
            print("Fizz")
        elif num%5==0:
            print("Buzz")
        elif num%3==0 and num%5==0:
            print("FizzBuzz")
        else:
            print(num)
        
        num+=1


print(print_fizzBizz())
    
```
